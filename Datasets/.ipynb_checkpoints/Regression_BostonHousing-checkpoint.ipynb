# Regression dataset
Here I load the Boston Housing dataset which predicts the house price in Boston from house details

```julia
using DataFrames
```

## 1. Load a dataset  as a dataframe and  convert it to a julia array

```julia
df = readtable("housing.data.txt", separator = ' ', header = false)
data = convert(Array, df)
```

## 2.  Get x and  y subset

```julia
x = data[:,1:end-1]
y = data[:,end]
```

## 3. Shuffel Divide into a training and a test subset

```julia
index = shuffle!(collect(1:size(y,1)))
i_train = index[1:round(Int,length(index)*0.6)]
i_test = index[round(Int,length(index)*0.6)+1:end]
y_train = y[i_train]
y_test = y[i_test]
x_train = x[i_train,:]
x_test = x[i_test,:]
```
